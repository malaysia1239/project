```python
import pandas as pd
import scipy.stats as stats

# Загрузка данных из файла
df = pd.read_csv('COVID-19 Survey Student Responses.csv')

# Перекодировка переменных на бинарные категории
df['normal_or_not_sleep'] = ['normal' if i > 7 else 'not normal' for i in df['Time spent on sleep']]
df['normal_or_not_social_media'] = ['normal' if i < 2 else 'not normal' for i in df['Time spent on social media']]

# Создание таблицы сопряженности
df_crosstab = pd.crosstab(df['normal_or_not_sleep'], df['normal_or_not_social_media'])

# Применение Хи-квадрат теста независимости
chi2, p, _, _ = stats.chi2_contingency(df_crosstab)

# Вывод результатов
print("Хи-квадрат статистика:", chi2)
print("p-значение:", p)
```

```python
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# Загрузка данных из файла
df = pd.read_csv('WVS Wave 6 2010-2014.csv')

# Стандартизация данных
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df)

# Применение метода главных компонент
pca = PCA(n_components=4)
pca_data = pca.fit_transform(scaled_data)

# Создание таблицы главных компонент и их факторных нагрузок
components_table = pd.DataFrame(pca.components_.T, columns=['PC1', 'PC2', 'PC3', 'PC4'])

# Определение переменных, попадающих в первый фактор
factor_loadings = components_table['PC1']
variables_in_factor = df.columns[np.abs(factor_loadings) >= 0.5]

# Вывод переменных, попадающих в первый фактор
print("Переменные, попадающие в первый фактор:", variables_in_factor.to_list())
```

```python
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

df = pd.read_csv('WVS Wave 6 2010-2014.csv')

scaler = StandardScaler()
scaled_data = scaler.fit_transform(df)

pca = PCA(n_components=4)
pca_data = pca.fit_transform(scaled_data)

# Вывод факторных нагрузок переменных
component_names = [f'PC{i+1}' for i in range(pca.n_components_)]
loadings = pd.DataFrame(pca.components_.T, columns=component_names, index=df.columns)

print(loadings)
```

```python
import pandas as pd
import numpy as np
from statsmodels.multivariate.factor import Factor
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
df = pd.read_csv('WVS Wave 6 2010-2014.csv')
for col in df.columns:
    df[col] = (df[col] - df[col].mean()) / df[col].std()
fa = Factor(df, n_factor=4, method='pa')
res = fa.fit()
res.get_loadings_frame()
```

```python
import statsmodels.api as sm
from statsmodels.genmod.generalized_linear_model import GLM
from statsmodels.genmod import families
from statsmodels.stats.outliers_influence import variance_inflation_factor
from sklearn.metrics import roc_auc_score, roc_curve
import pandas as pd

df = pd.read_csv('loan.csv')

model_5 = sm.GLM(df["Loan_Status"], df[['LoanAmount', 'Married', 'Education']], family=families.Binomial()).fit()

prob_pred = model_5.predict()

fpr, tpr, thresholds = roc_curve(pd.to_numeric(df["Loan_Status"]), prob_pred)
roc_auc = roc_auc_score(pd.to_numeric(df["Loan_Status"]), prob_pred)

print("ROC AUC Score:", roc_auc)
```
